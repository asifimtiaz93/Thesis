```python
file_path_train = ('/content/drive/MyDrive/Colab Notebooks/twitter_train.csv')
df_train = pd.read_csv(file_path_train)
df_train.head()
```

```python
# prompt: Take only text and target column form df_train form above cell. Then run count vectorzation

from sklearn.feature_extraction.text import TfidfVectorizer

# Assuming 'text' and 'target' are column names in your DataFrame
X = df_train['text']
y = df_train['target']

vectorizer = TfidfVectorizer()
X_vectorized = vectorizer.fit_transform(X)

X_vectorized.shape
```

```python
# prompt: now feed this to multinomailNB

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_vectorized, y, test_size=0.2, random_state=42)

# Initialize and train the Multinomial Naive Bayes classifier
clf = MultinomialNB()
clf.fit(X_vectorized,y)

# Make predictions on the test set
y_pred = clf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")


# Now you can use the trained classifier to predict on new text data
# Example:
new_text = ["This is a sample tweet."] # Replace with your new text data

# Vectorize the new text using the same vectorizer
new_text_vectorized = vectorizer.transform(new_text)

# Predict the target variable
predicted_target = clf.predict(new_text_vectorized)

print(f"Predicted target for new text: {predicted_target}")
```
